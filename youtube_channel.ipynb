```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from webdriver_manager.chrome import ChromeDriverManager
import time
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings("ignore")
```

```python
driver=webdriver.Chrome(ChromeDriverManager().install())
driver.get("https://youtube.com")# url of site
time.sleep(3)
channel=driver.find_element(By.CSS_SELECTOR,'INPUT[id="search"]').send_keys("The cooking art n craft house")#write a name 
time.sleep(3)
channelbutton=driver.find_element(By.CSS_SELECTOR,'button[id="search-icon-legacy"]').click()#search name
time.sleep(3)
channel_info=driver.find_element("xpath",'//*[@id="content-section"]').click()#select channel
time.sleep(3)
cahnnel_video=driver.find_element("xpath",'//*[@id="tabsContent"]/tp-yt-paper-tab[2]').click()#goto videos
time.sleep(3)
hover_page=driver.find_element(By.TAG_NAME,'body')#page body
hover_page.send_keys(Keys.PAGE_DOWN)#go to end of the page

for i in range(1,90):#number o pages
    hover_page.send_keys(Keys.PAGE_DOWN)#to reach page end
    
vide=driver.find_elements(By.CLASS_NAME,"style-scope ytd-rich-grid-media")#go to  video
video_list=[] #create a list to store values


for info in vide:
    title=info.find_element(By.XPATH,".//*[@id='video-title']").text # used to scrape data of title  
    print(title)
    views=info.find_element(By.XPATH,'.//*[@id="metadata-line"]/span[1]').text # used to scrape data of views 
    print(views)
    time=info.find_element(By.XPATH,'.//*[@id="metadata-line"]/span[2]').text
    print(time)
    video_info={
    'Title':title,
    'Views': views,
    'Time_upload':time
    }
    video_list.append(video_info)
```

```python
df=pd.DataFrame(video_list)
df
```

```python
df.to_csv("yotube channel.csv")
```
